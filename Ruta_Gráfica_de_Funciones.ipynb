### Elementos a tener en cuenta para Analizar la grafica de una función

* Una función es una relación entre dos variables x e y, de modo que, a cada valor de la variable independiente, x, le asocia un único valor de la variable  dependiente y.
* El dominio de una función es el conjunto de todos los posibles valores que puede tomar x.
* La gráfica de una función es el conjunto de puntos $ (x,f(x))$  representados en el plano.
* Una función es continua si puede representarse con un solo trazo. Es discontinua en un punto si presenta un "salto", un "huequito" o asíntota.
* Una función es simétrica respecto al eje Y, o  función par, si $f(x)=f(-x)$; y es simétrica respecto al origen, función impar, si $f(-x)=-f(x)$
* Una función es creciente en un intervalo, cuando dados dos puntos cualesquiera del mismo Si $ x_1<x_2$  entonces $f(x_1)<f(x_2)$ y decreciente si $x_1<x_2$ entonces $ f(x_1)>f(x_2)$

```python
import matplotlib.pyplot as plt
from ipywidgets import interact

def graficar_lineal(pendiente):
    x = range(-10, 11)
    y = [pendiente * xi for xi in x]

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'Pendiente: {pendiente}', color='blue')
    plt.title('Función Lineal')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.show()

# Utilizando ipywidgets para crear un control deslizante interactivo
interact(graficar_lineal, pendiente=(-10, 10, 0.5))
```

# Grafica de la función lineal

Explore dando valores a la pendiente y el intercepto y reflexione sobre el comportamiento de la grafica de la linea

```python
# @title FUNCIÓN LINEAL  { run: "auto", vertical-output: true }
pendiente = 4 # @param {type:"number"}
intercepto = 4 # @param {type:"number"}

# Crear datos x e y para la línea
x = range(-10, 11)  # Puedes ajustar el rango según tus necesidades
y = [pendiente * i + intercepto for i in x]

# Graficar la línea
plt.plot(x, y, label=f'Línea: y = {pendiente}x + {intercepto}')

# Agregar etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.axvline(x=0, color='black') #Eje y
plt.axhline(y=0, color='black') #Eje x
plt.title('Gráfica de una línea')

# Mostrar la leyenda
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

# Recurso Adicional

Observe lo que sucede en el grafico al variar los valores de a,b y c en la función cuadrática

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

def graficar_cuadratica(a, b, c):
    x = np.linspace(-10, 10, 100)
    y = a * x**2 + b * x + c

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'{a}x^2 + {b}x + {c}', color='red')
    plt.title('Función Cuadrática')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()

#Creando el deslizador
interact(graficar_cuadratica, a=(-5, 5, 0.1), b=(-5, 5, 0.1), c=(-5, 5, 0.1))
```

# ¿Qué es un modelo?

**Un Modelo Matemático**: Es una expresión o función matemática usada para describir una situación u objeto real.

Para modelar funciones, algunos expertos recomiendan tener en cuenta las siguientes consideraciones:

•	Identifique la variable independiente del problema.  Usualmente es el tiempo, el precio de un producto, una longitud en metros, etc.
•	Identifique las constantes y las otras variables del problema.
•	Exprese todas las variables en términos de la variable independiente, para ello utilice geometría o las relaciones propias del contexto del problema.
•	Construya una expresión en términos de las variables para el problema que se va a modelar.
•	Determine el dominio de la función, recuerde que en estos casos el dominio depende también del contexto del problema.
